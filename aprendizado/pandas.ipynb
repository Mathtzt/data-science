```python
# DataFrames
# note que é possível criar as próprias linhas e colunas, assim como configurá-las
```

```python
data = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
data
```

```python
df1 = pd.DataFrame(data)
df1
```

```python
index = ['Linha' + str(i) for i in range(df1.shape[0])]
index
```

```python
df1 = pd.DataFrame(data, index)
df1
```

```python
columns = ['Coluna' + str(i) for i in range(df1.shape[0])]
df1 = pd.DataFrame(data, index, columns)
df1
```

```python
df1[df1 > 0] = 'A'
df1
```

```python
df2 = df1.copy()
df2[df2 == 'A'] = 'B'
df2
```

```python
## É possível concatenar DataFrames passando uma lista de dataframes par ao método concat()
# Pelas colunas
df3 = pd.concat([df1, df2])
df3
```

```python
# Pelas linhas
df4 = pd.concat([df1, df2], axis=1)
df4
```

### Parte 4

```python
list(dados['Tipo'].drop_duplicates())
```

```python
residencial = ['Quitinete',
 'Casa',
 'Apartamento',
 'Casa de Condomínio',
 'Flat',
 'Casa de Vila',
 'Loft',
 'Chácara',
 'Loteamento/Condomínio',
 'Sítio']
```

```python
residencial
```

```python
selecao = dados['Tipo'].isin(residencial)
selecao.head(10)
# método isin() verifica dentro do DataFrame se os elementos da nossa lista existem de fato, assinalando True ou False.
```

```python
dados_residencial = dados[selecao]
# quando passamos uma selação uma series booleana para o dataframe, ele mostrará apenas as tuplas que são True
```

```python
dados_residencial
```

```python
list(dados_residencial['Tipo'].drop_duplicates())
# comprovando que o novo data frame só tem os imóveis do tipo residencial
```

```python
dados_residencial.index = range(dados_residencial.shape[0])
# reconstruindo os index
dados_residencial
```

```python
## exportando a base de dados
```

```python
dados_residencial.to_csv("aluguel_residencial", sep=";", index=False)
# método to_csv gera um arquivo csv.
# note que o primeiro param é o path de destino
# o attr index false remove a indexação na geração do arquivo. Isso é importante para que não tenha uma indexação duplicada quando o arquivo for lido
```
