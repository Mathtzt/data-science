```python
# Importações
import pandas as pd

from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
uri = "https://gist.githubusercontent.com/Mathtzt/09d8642eb5c4eca855f47771cd3393a3/raw/9026412c10ce6bab95341602177fae71ff41d469/tracking_web_dataset.csv"
```

```python
# Leitura dos dados
dados = pd.read_csv(uri, sep=",")
dados.head()
```

```python
dados.shape
```

```python
# Separando variáveis explicativas da variável dependente.
x = dados.iloc[:, 0:3]
y = dados.iloc[:, 3]
```

```python
# Separando dataset de treino e teste inicialmente na mão
x_train = x[:75]
y_train = y[:75]
x_test = x[75:]
y_test = y[75:]

print("Treinaremos com %d elementos, e testaremos com %d elementos" % (len(x_train), len(x_test)))
```

```python
# Separando dataset de treino e teste utilizando o sklearn
seed = 20
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = seed)
print("Treinaremos com %d elementos, e testaremos com %d elementos" % (len(x_train), len(x_test)))
```

```python
# Criando modelo
model = LinearSVC()

# Treinando o modelo
model.fit(x_train, y_train)

# Prevendo com dados de teste
y_pred = model.predict(x_test)
```

```python
# Avaliando nosso modelo
accuracy = accuracy_score(y_test, y_pred)
print("A acurácia do modelo = %.2f%%" % (accuracy * 100))
```

Obs: a proporção das classes de treino é diferente das de teste. Isso pode inviesar nosso modelo

```python
# Separando dataset de treino e teste mantendo a proporção da nossa classe dependente
# Para tal utilizamos o parâmetro stratify
seed = 20
x_train, x_test, y_train, y_test = train_test_split(x, 
                                                    y, 
                                                    test_size = 0.25, 
                                                    random_state = seed,
                                                    stratify = y)
print("Treinaremos com %d elementos, e testaremos com %d elementos" % (len(x_train), len(x_test)))
```

```python
y_train.value_counts()
```

```python
y_test.value_counts()
```

```python
# Criando modelo
model = LinearSVC()

# Treinando o modelo
model.fit(x_train, y_train)

# Prevendo com dados de teste
y_pred = model.predict(x_test)

# Avaliando nosso modelo
accuracy = accuracy_score(y_test, y_pred)
print("A acurácia do modelo = %.2f%%" % (accuracy * 100))
```
