```python
"""
Features:
- Pelo longo
- Perna curta
- Faz 'au au'
"""
porco1 = [0, 1, 0]
porco2 = [0, 1, 1]
porco3 = [1, 1, 0]

cachorro1 = [0, 1, 1]
cachorro2 = [1, 0, 1]
cachorro3 = [1, 1, 1]

X_train = [porco1, porco2, porco3, cachorro1, cachorro2, cachorro3]

"""
Classes:
- 0 = cachorro
- 1 = porco
"""
y_train = [1, 1, 1, 0, 0, 0] #labels
```

```python
# Importando Sklearn
# De forma introdutória, será importado o estimador LinearSVC
from sklearn.svm import LinearSVC
```

```python
# Criando modelo
model = LinearSVC()
# Treinando modelo
model.fit(X_train, y_train)
```

```python
# Teste simples
animal_misterioso = [1, 1, 1]
result = model.predict([animal_misterioso])

if result[0] == 0:
  print("É um cachorro!")
else:
  print("É um porco")
```

```python
# Testes com outros casos
misterio1 = [1, 1, 1]
misterio2 = [1, 1, 0]
misterio3 = [0, 1, 1]

X_test = [misterio1, misterio2, misterio3]
y_test = [0, 1, 1]
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
# Importando metrics para validar o modelo
from sklearn.metrics import accuracy_score
```

```python
taxa_de_acerto = accuracy_score(y_test, y_pred)
print("Taxa de acerto(acurácia) = %.2f%%" % (taxa_de_acerto * 100))
```
